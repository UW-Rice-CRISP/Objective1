```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.corpus.reader import PlaintextCorpusReader
import string
import re # regular expression library
import csv
import sys
import pandas as pd
from __future__ import division

%pylab inline
```

```python
# read in corpus within folder "./corpus". Each corpus file is tokenized at both the sentence and word level.
# See NLTK docs to see how to access tokens ... e.g., corpus.words('file.txt')

folder = '../TextData/CanterburyNewsStoriesCleaned/'
corpus = PlaintextCorpusReader(folder, '.*\.txt',encoding="utf8")
#corpus.fileids()
```

```python
# function to clean words of stop words and punctuation.

def clean_words(words):
	punctuations = list(string.punctuation) + ["''",'...']

	# multi-word strings only there to keep track of what phrases were manually deleted from files
	my_stop_words = []
	
	stop_words = stopwords.words('english') + my_stop_words + punctuations


	return [w.lower() for w in words if w.lower() not in stop_words]
```

```python
# create a file of collocations within each corpus file by looping through each filename

filename = 'collocations.txt'

# Open the file
filecolloc = open(filename, 'w')
# Save old stdout stream
tmpout = sys.stdout
# Redirect all "print" calls to that file
sys.stdout = filecolloc

for fileid in corpus.fileids():
    words = corpus.words(fileid)
    words_clean = clean_words(words)
    text = nltk.Text(words_clean)
    print '\n', fileid
    text.collocations()
    
# Close file
filecolloc.close()
# Reset stdout in case you need something else to print
sys.stdout = tmpout
```

```python
# create a file of concordances within each corpus file by looping through each filename

filename = 'concordances.txt'
concord_word = 'resilience' # word to find concordances for
concord_lines = 1200 # number of output lines/concordances
concord_context = 100 # number of characters extracted from left and right of concord_word

# Open the file
fileconcord = open(filename, 'w')
# Save old stdout stream
tmpout = sys.stdout
# Redirect all "print" calls to that file
sys.stdout = fileconcord

for fileid in corpus.fileids():
    words = corpus.words(fileid)
    text = nltk.Text(words)
    print '\n', fileid
    text.concordance(concord_word, concord_context, concord_lines)
    
# Close file
fileconcord.close()
# Reset stdout in case you need something else to print
sys.stdout = tmpout       
```

```python
# can find each word in corpus that starts with the prefix "infra" using regular expression
[w for w in corpus.words() if re.search('infra$', w)]
```

```python
# can find each word in corpus that ends with the suffix "structure" using regular expression
[w for w in corpus.words() if re.search('^structure', w)]
```

```python
# can find each word in corpus that contains "elect" using regular expression
[w for w in corpus.words() if re.search('*elect*', w)]
```

```python
# create a frequency distribution of words within the corpus (or can indicate single file) and plot first 40 words
fd = nltk.FreqDist(clean_words(corpus.words()))

rcParams['axes.labelsize'] = '20'
tick_params(axis='both', labelsize=20)
figsize(20,10)

fig = fd.plot(40,cumulative=False)
```

```python
# can throw frequency distribution into a pandas dataframe to do further analysis
corpus_freq_table = pd.DataFrame(fd.items())
corpus_freq_table.sort(columns=1, axis=0, ascending=False, inplace=True, kind='quicksort', na_position='last')
corpus_freq_table.columns = ['Word', 'Count']
corpus_freq_table.reset_index(drop=True, inplace=True)
corpus_freq_table[:5]
```

```python
# for example

fig, ax = subplots()


corpus_freq_table[:25].plot(ax=ax, x='Word',y='Per Thousand',legend=False,rot=90, kind='bar')
```

```python
# create a dispersion diagram of the corpus based on given words

phase_words = ['preparedness','response','mitigation','restoration','recovery',
              'time','days','weeks','months','years']

disp_title = ''

fig, ax = subplots()
rcParams['axes.labelsize'] = '20'
tick_params(axis='both', labelsize=20)
figsize(20,5)
nltk.draw.dispersion_plot(clean_words(corpus.words()),phase_words,
                         title=disp_title)
```
